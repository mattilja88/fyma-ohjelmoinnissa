```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

data = pd.read_csv("./kiihtyvyysdata/rawdata.csv")

time = data["Time (s)"]
ax = data["Linear Acceleration x (m/s^2)"]
ay = data["Linear Acceleration y (m/s^2)"]
az = data["Linear Acceleration z (m/s^2)"]
absolute = data["Absolute acceleration (m/s^2)"]

fig, axs = plt.subplots(5, 1, figsize=(8, 10), sharex=True)

axs[0].plot(time, ax, color="r")
axs[0].set_ylabel("a (m/s)")
axs[0].set_title("Linear Acceleration X")

axs[1].plot(time, ay, color="g")
axs[1].set_ylabel("a (m/s)")
axs[1].set_title("Linear Acceleration Y")

axs[2].plot(time, az, color="b")
axs[2].set_ylabel("a (m/s)")
axs[2].set_title("Linear Acceleration Z")

axs[3].plot(time, absolute, color="b")
axs[3].set_ylabel("a (m/s)")
axs[3].set_title("Absolute acceleration")

axs[4].plot(time, ax, label="ax", color="r")
axs[4].plot(time, ay, label="ay", color="g")
axs[4].plot(time, az, label="az", color="b")
axs[4].plot(time, absolute, label="absolute", color="k", linestyle="--")
axs[4].set_xlabel("t (s)")
axs[4].set_ylabel("a (m/s)")
axs[4].set_title("Multi")
axs[4].legend()

plt.tight_layout()

pdf_name = "kiihtyvyyskuvaajat.pdf"
with PdfPages(pdf_name) as pdf:
    pdf.savefig(fig, bbox_inches="tight")
    
plt.show()
```
